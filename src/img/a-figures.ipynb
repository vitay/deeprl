```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns 

sns.set(style="white", palette="muted", color_codes=True)
sns.set_context("talk")
```

## Scheduling

```python
x = np.linspace(0, 1000, 1001)

y = 0.3 * (900 - x)/1000
y[y<0.05] = 0.05
z = 0.3 * (np.exp(-(4*x)/1000))

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.plot(x, y)
plt.ylim((0.0, 0.3))
plt.xlabel("Steps")
plt.ylabel("Exploration parameter")
plt.title("Linear scheduling")
plt.subplot(122)
plt.plot(x, z)
plt.xlabel("Steps")
plt.ylabel("Exploration parameter")
plt.title("Exponential scheduling")

sns.despine()
plt.tight_layout()
plt.savefig("bandit-scheduling.png")
```
